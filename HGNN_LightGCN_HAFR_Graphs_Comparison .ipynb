```python
import matplotlib.pyplot as plt

# Metrics for each model
models = ['SCHGN', 'LightGCN', 'HAFR']
auc = [0.6333, 0.6340, 0.6314]
recall_10 = [0.0784, 0.0892, 0.0734]
recall_20 = [0.1323, 0.1494, 0.1297]
ndcg_10 = [0.0741, 0.0707, 0.0668]
ndcg_20 = [0.0907, 0.0896, 0.0856]

# Plotting individual graphs for each metric
plt.figure(figsize=(14, 10))

# AUC
plt.subplot(3, 2, 1)
plt.plot(models, auc, marker='o', label='AUC')
plt.title('AUC')
plt.xlabel('Models')
plt.ylabel('Score')
plt.grid(True)

# Recall@10
plt.subplot(3, 2, 2)
plt.plot(models, recall_10, marker='o', label='Recall@10')
plt.title('Recall@10')
plt.xlabel('Models')
plt.ylabel('Score')
plt.grid(True)

# Recall@20
plt.subplot(3, 2, 3)
plt.plot(models, recall_20, marker='o', label='Recall@20')
plt.title('Recall@20')
plt.xlabel('Models')
plt.ylabel('Score')
plt.grid(True)

# NDCG@10
plt.subplot(3, 2, 4)
plt.plot(models, ndcg_10, marker='o', label='NDCG@10')
plt.title('NDCG@10')
plt.xlabel('Models')
plt.ylabel('Score')
plt.grid(True)

# NDCG@20
plt.subplot(3, 2, 5)
plt.plot(models, ndcg_20, marker='o', label='NDCG@20')
plt.title('NDCG@20')
plt.xlabel('Models')
plt.ylabel('Score')
plt.grid(True)

# Recall and NDCG combined
plt.subplot(3, 2, 6)
plt.plot(models, recall_10, marker='o', label='Recall@10')
plt.plot(models, recall_20, marker='o', label='Recall@20')
plt.plot(models, ndcg_10, marker='o', label='NDCG@10')
plt.plot(models, ndcg_20, marker='o', label='NDCG@20')
plt.title('Recall and NDCG Combined')
plt.xlabel('Models')
plt.ylabel('Score')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
